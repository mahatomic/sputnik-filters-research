```python
import pandas as pd
```

Загружаем датасет

```python
df = pd.read_csv('/content/events filters к тестовому Product Analyst Q1 2024.csv')
```

```python
df
```

```python
df.info()
```

Из event_label выделяем города в отдельный столбец city и информацию, куда кликнул пользователь (после /) в столбец info

```python
df['city'] = df['event_label'].str.split('/').str[0].str.strip()
df['info'] = df['event_label'].str.split('/').str[1].str.strip()
df
```

Группируем по действиям с подсчетом общего количества всех и уникальных событий

```python
grouped_on_action_df = df.groupby('event_action')[['total_events', 'unique_events']].sum().reset_index()
grouped_on_action_df
```

Общее количество событий по действию search-tools-button_open больше, чем
по действиям filters-categories_click	и price_button_submit, которые указывают на применение фильтра

```python
search_df = df[df['event_action'] == 'search-tools-button_open']
search_df.head(100)
```

Фактически действие search-tools-button_open - это просто открытие фильтра, но не обязательно его использование в контексте применения, поэтому в расчете это действие учитывать не будем

```python
filtered_df = df[df['event_action'] == 'filters-categories_click'].copy()
filtered_df.head(100)
```

В действии filters-categories_click отражено использование пользователем сортировок и категорий в рамках фильтрации

Распределим использование категорий и сортировок по разным датафреймам и удалим из расчета данные, обозначенные капсом, так как предполагаю, что это грязные данные, которые не имеют отношения к фильтрации

Удаляем строчки, где встречается текст капсом

```python
filtered_df = filtered_df[~filtered_df['info'].str.isupper()]
filtered_df
```

Выделим сначала в отдельный датафрейм события использования сортировок и подсчитаем их общее количество

```python
phrases = [
    'Сначала самые дешевые',
    'Сначала самые дорогие',
    'Сначала самые популярные',
    'По рейтингу'
]
or_phrases = '|'.join(phrases)
sort_df = filtered_df[filtered_df['info'].str.contains(or_phrases, case=False, na=False)].copy()
sort_df
```

```python
total_sort = sort_df[['total_events', 'unique_events']].sum()
res_sort = total_sort.to_frame('total_sum')
res_sort
```

Получаем, что сортировка была использована 20430 раз, из которых 17382 раз уникальными пользователями

Для создания отдельного датафрейма с категориями удалим из filtered_df строки, которые встречаются в total_sort

```python
cat_df = filtered_df[~filtered_df.index.isin(sort_df.index)]
cat_df
```

```python
total_cat = cat_df[['total_events', 'unique_events']].sum()
res_cat = total_cat.to_frame('total_sum')
res_cat
```

Таким образом, категории были выбраны 57835 раз, из которых 43863 раз уникальными пользователями

Далее рассмотрим применение фильтров. В расчете не будем учитывать действия clear_filter_mobile	и price_button_close, так как эти действия скорее относятся не к фильтрации, а к ее завершению или прерыванию

Судя по особенностям работы сайта, фильтр применяется сразу при нажатии на вариант из предложенного списка на чекбоксе, и при этом не обязательно нажимать "Показать"

Поэтому в расчете также не будем учитывать действие price_button_submit, но при этом просуммируем количество событий всех возможных вариантов применения фильтров

```python
filter_actions = [
    'dates_filter_mobile',
    'end_date_click',
    'pay-type_checkbox',
    'price_first',
    'price_second',
    'price_third',
    'start_date_click',
    'ticket-type_checkbox'
]
filt_df = df[df['event_action'].isin(filter_actions)]
filt_df
```

```python
total_filt = filt_df[['total_events', 'unique_events']].sum()
res_filt = total_filt.to_frame('total_sum')
res_filt
```

Видим, что общее количество событий использования фильтрации 90120, при этом от уникальных пользователей - 69160

Теперь расчитаем сколько людей пользуются фильтрами и оценим, какие разделы фильтров наиболее востребованы

```python
total_all = pd.DataFrame({
    'Фильтры': total_filt,
    'Категории': total_cat,
    'Сортировки': total_sort
}).T
total_all.loc['Всего'] = total_all[['total_events', 'unique_events']].agg('sum')
total_all
```

```python
total_unique_events = total_all.loc['Всего', 'unique_events']
unique_page_visits = grouped_on_action_df[grouped_on_action_df['event_action'] == 'Page Visit']['unique_events']
uniq_total = {
    'total_unique_events': total_unique_events,
    'share_of_uniq_page_visits (%)': (total_unique_events / unique_page_visits * 100). round(2),
}
unique__df = pd.DataFrame(uniq_total)
unique__df
```

Можно сделать вывод, что 130405 уникальных пользователей пользуются фильтрами всех категорий, что составляет 9,42% от количества уникальных посещений сайта

Рассмотрим распределение уникальных пользователей между разделами фильтров

```python
unique_data = total_all['unique_events'][:-1]
percentages = (unique_data / unique_data.sum() * 100).round(2)
unique_table = pd.DataFrame({
    'unique_events': unique_data,
    'share (%)': percentages
})
unique_table
```

Наиболее востребован раздел "фильтры" (53,03%), далее идут "категории" (33,03%), и наименее популярны "сортировки" (13,33%)

Далее посмотрим, как пользуются фильтрами по городам

```python
all_filrters = pd.concat([
    filt_df,
    cat_df,
    sort_df
])
city_group = all_filrters.groupby('city', as_index=False)[['total_events', 'unique_events']].sum()
city_group
```

```python
sorted_city_group = city_group.sort_values(['total_events', 'unique_events'], ascending=[False, False])
sorted_city_group
```

В результате получаем, что в топ 5 городов, где используются фильтры, входят Калининград, Санкт-Петербург, Москва, Сочи и Стамбул

```python
sorted_city_group.tail(50)
```

Городов, где пользуются фильтрами меньше всего, достаточно много, при этом очень много единичного использования. К примеру, такими городами являются Саранск, Сингапур, Сургут, Тунис, Виченца, Висбаден

Рассмотрим, как часто люди пользуются фильтром выбора цены

```python
price_filters = ['price_first', 'price_second', 'price_third']
price_df = grouped_on_action_df[grouped_on_action_df['event_action'].isin(price_filters)].copy()
price_df
```

Оставим только данные для уникальных пользователей

```python
price_df = price_df.drop(columns=['total_events'])
price_df
```

Посчитаем общее количество использований фильтров выбора цены

```python
price_df = price_df.reset_index(drop=True)
price_unique_events = price_df['unique_events'].sum()
price_df.loc[len(price_df)] = ['Всего', price_unique_events]
price_df
```

```python
price_df['share (%)'] = ((price_df['unique_events'] / price_unique_events) * 100).round(2)
price_df[:-1]
```

Больше всего пользователи пользуются единичным применением фильтра цены (72,05%) и меньше всего трехкратным (6,42%), что вполне логично

```python
total_filter_events = total_all.loc['Фильтры', 'unique_events']
data_uniq_price = {
    'price_unique_events': [price_unique_events],
    'share_unique_filt (%)': [((price_unique_events / total_filter_events) * 100). round(2)],
    'share_unique_filt_all (%)': [((price_unique_events / total_unique_events) * 100).round(2)]
}
unique_price_df = pd.DataFrame(data_uniq_price)
unique_price_df
```

Итак, фильтр выбора цены использовался пользователями 9266 раз, что составляет 13,4% от количества событий использования раздела "фильтры" и 7,11% от количества событий использования в принципе фильтров (всех разделов)

Наконец, любопытно посмотреть, какие фичи из раздела "фильтры" используются пользователями чаще всего, а какие меньше всего

```python
features_df = grouped_on_action_df[grouped_on_action_df['event_action'].isin(filter_actions)]
features_df
```

```python
features_df = features_df.drop(columns=['total_events']).reset_index(drop=True)
```

Рассчитаем долю использования фичей уникальными пользователями относительно общего количества использования раздела "фильтры" и расположим их по убыванию

```python
features_df['share_of_filters (%)'] = (features_df['unique_events'] / total_filter_events * 100).round(2)
features_df = features_df.sort_values('share_of_filters (%)', ascending=False)
features_df
```

Из расчетов видно, что в разделе "фильтры" чаще всего уникальные пользователи пользуются фильтром на размер группы (25,91%), а далее идут фильтры по датам (начало, конец и сегодня/завтра). Меньше всего пользователи задействуют фильтры на повторное изменение стоимости
